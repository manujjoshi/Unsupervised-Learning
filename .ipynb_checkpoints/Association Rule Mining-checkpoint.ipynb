# `Association Rule Mining -------------------------`
- `pick out frequent items which are bought together, called Market Basket Analysis, It analyses basket of a customers and find which of the products are bought together`
- `If a human being tend to find the pattern, the seller will change the pattern`

```python
import pandas as pd
```

```python
data = pd.read_csv("Market_Basket_Optimisation.csv",header=None)
```

```python
data
```

### `data processing(Converting into transaction),I don't want them as columns,want them as a list`

### `one row will be one list`

```python
unique_item = []
for col in data.columns:
    unique_item.extend(data[col].unique().tolist())
```

### `unique items in my entire transactions`

```python
unique_item
```

### `remove nan`

```python
items = list(set([i for i in unique_item if str(i)!="nan"]))
items
```

### `prepare transaction`

```python
encoded_tranaction = []
for index,row in data.iterrows():  # iterate over dataframe
    
    labels = {}
    item_in_row = row
    uncommon = set(items)- set(item_in_row)
    common = set(items) & set(item_in_row)
    
    for i in uncommon:
        labels[i]=0
        
    for i in common:
        labels[i]=1
        
    encoded_tranaction.append(labels)
```

```python
encoded_tranaction
```

```python
transaction_df  = pd.DataFrame(encoded_tranaction)
```

```python
transaction_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(24,8))
transaction_df.sum().plot.bar()
plt.show()
```

```python
# !pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
# apriori => frequent item sets
# association_rules => rules
```

```python
freq_items = apriori(transaction_df,min_support=0.01,max_len=4,use_colnames=True)
```

```python
freq_items = freq_items.sort_values(ascending=False,by='support')
freq_items
```

### `frequent item sets will only have support, rules will only have confidence`

### `first 50 ( support vs itemsets )`

```python
freq_items.iloc[0:50,:].plot.bar(x="itemsets",y="support",figsize=(24,8))
plt.show()
```

### `prepare association rules which will be used for my recommendation`

```python
rules = association_rules(freq_items,metric="confidence",min_threshold=0.1)
```

```python
rules
```

### `I just want lift greater than 1`

```python
rules = rules[rules["lift"]>1].sort_values("lift",ascending=True)
```

```python
rules
```

### `Recommendation system for one product ----------------------------------`

```python
customer_item = "mineral water"
rules[rules["antecedents"].apply(str).str.contains(customer_item)].sort_values("confidence",ascending=False)
```
